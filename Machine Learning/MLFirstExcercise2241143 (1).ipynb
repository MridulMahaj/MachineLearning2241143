```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load dataset
data = pd.read_csv('MBA.csv')

gpa = data['gpa']

# Define bins for grouping GPA values
bins = [0, 2, 2.6, 2.8, 3.2, 3.4, 3.6, 3.8, 4]  # Adjusted based on your dataset
gpa_bins = pd.cut(gpa, bins=bins, right=False)

# Count the number of GPAs in each bin
gpa_counts = gpa_bins.value_counts().sort_index()

# Plot bar graph
plt.bar(gpa_counts.index.astype(str), gpa_counts, color='skyblue')
plt.title('GPA Distribution (Bar Graph)')
plt.xlabel('GPA Range')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()


# Calculate mean and standard deviation
mean = gpa.mean()
std_dev = gpa.std()

# Generate data for the normal distribution curve
x = np.linspace(gpa.min(), gpa.max(), 100)
y = norm.pdf(x, mean, std_dev)

# Plot the histogram for GPA
plt.hist(gpa, bins=10, density=True, alpha=0.6, color='skyblue', label='Histogram')

# Plot the normal distribution curve
plt.plot(x, y, color='red', label='Normal Distribution')

# Add titles and labels
plt.title('GPA Normal Distribution')
plt.xlabel('GPA')
plt.ylabel('Density')
plt.legend()
plt.show()

#We have plotted the GPA distribution in the form of a bar graph
```

```python
#Bivariate Analysis for columns GPA and GMAT score
#Scatterplot for correlation mapping
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset
# Replace 'your_dataset.csv' with your actual file name
data = pd.read_csv('MBA.csv')

# Extract GPA and GMAT Score columns
x = data['gpa']  # Independent variable
y = data['gmat']  # Dependent variable

# Scatter plot
plt.scatter(x, y, label="Data points")

# Fit line
coefficients = np.polyfit(x, y, 1)  # 1st-degree polynomial (linear fit)
poly = np.poly1d(coefficients)

# Plot the best-fit line
plt.plot(np.unique(x), poly(np.unique(x)), color='red', label="Best-fit line")

# Add titles and labels
plt.title('Correlation between GPA and GMAT Score')
plt.xlabel('GPA')
plt.ylabel('GMAT Score')

# Display plot
plt.show()

#The correlation between GMAT and GPA socre is a positive one according to the scatter plot
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
mba = pd.read_csv("MBA.csv")

# Clean the dataset (if necessary)
# Replace missing values in numeric columns with their mean
for col in mba.select_dtypes(include=['number']).columns:
    mba[col].fillna(mba[col].mean(), inplace=True)

# Scatterplot for analysis
plt.figure(figsize=(8, 6))
sns.scatterplot(x='gmat', y='gpa', data=mba)

# Add title and labels
plt.title('Scatterplot of GMAT vs GPA')
plt.xlabel('GMAT Score')
plt.ylabel('GPA')

# Show the plot
plt.show()

#We have described the comparison between 2 columns using GMAT and GPA with a scatterplot
```

```python
#Multivariate Analysis using a HeatMap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
mba = pd.read_csv("MBA.csv")

# Clean the dataset
if 'Gender' in mba.columns: 
    mba['Gender'] = mba['Gender'].map({'Male': 1, 'Female': 0})

# Handle missing values
# Replace missing values in numeric columns with the column mean
for col in mba.select_dtypes(include=['number']).columns:
    mba[col].fillna(mba[col].mean(), inplace=True)

# Ensure only numeric columns for correlation
numeric_data = mba.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlations')
plt.show()

#We have different correlations such as negative correlation between gpa and work experience
#Both columns GPA and GMAT have a negative correlation with respect to work experience
```
