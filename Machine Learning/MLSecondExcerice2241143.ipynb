```python
#For missing values
# Handling missing values in 'race' using mode and 'admission' using a placeholder
import pandas as pd
data = pd.read_csv("MBA.csv")
data['race'].fillna(data['race'].mode()[0], inplace=True)
data['admission'].fillna('Not Admitted', inplace=True)

# Displaying the dataset summary post-missing data treatment
print(data.info())

# Interpretation: Missing values in 'race' replaced with the mode, and 'admission' missing values replaced with 'Not Admitted'.
```

```python
#Distribution of numeric variables using a histogram
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_columns:
    sns.histplot(data=data, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Interpretation: The histograms show the distribution and density of numeric variables, identifying their spread and skewness.

'''The GPA distribution reflects a bell-shaped curve, suggesting that most students perform consistently within a moderate range. The majority of 
GPAs cluster between 3.2 and 3.4, indicating that students are achieving well above average, but outliers on either side of the spectrum are uncommon. 
This kind of distribution is typical in academic settings, where performance tends to center around a common average.

For GMAT scores, the distribution leans slightly to the right, with most scores concentrated between 600 and 700. This shows that the applicant pool 
is strong, with many achieving respectable scores, but theres still a small number of individuals excelling with scores closer to 750. 
The slight skew indicates a diverse range of test-takers, with few scoring exceptionally low or exceedingly high.

The work experience data tells a slightly more nuanced story. While the clear peak at 5 years shows that most applicants are in the sweet spot of 
mid-career experience, smaller peaks at 4 and 6 years hint at a dynamic applicant pool. Very few individuals have less than 3 years or more than 8 years
of experience, suggesting that the program attracts professionals who have gained significant, yet manageable, career exposure before applying.'''
```

```python
#Distribution of nominal variables using count plot
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
categorical_columns = data.select_dtypes(include=['object', 'bool']).columns

for col in categorical_columns:
    sns.countplot(data=data, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()

# Interpretation: Count plots provide insights into the frequency distribution of categorical variables.
# For gender, females have a lower count(150-200) rather than Male(above 300) proving that males are more in number in MBA
# For internationnal column, most of the students are not international, ie most of them are Americans
# For distributon of major, we see that Humnities has the largest count
# For race column, we see that Whites are more in number with Asians just behind
# For work industry, we see that Investment Manageology has the largest count
# For admissions column, we see that most of these students have been admitted while very few are in the waitlist
```

```python
# 6. Comparison of numeric variables using scatterplot
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
sns.scatterplot(data=data, x='gmat', y='gpa', hue='admission')
plt.title('GMAT vs GPA')
plt.show()

# Interpretation: Scatterplot reveals the relationship between GMAT scores and GPA and highlights trends across admission statuses.
# We see that most of the students who have a high GPA and high GMAT scoes have been admitted into MBA
```

```python
# 7. Heatmap for correlation
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
# Filter numeric columns for correlation
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()

# Generate the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Interpretation: The heatmap shows relationships among numeric variables in the dataset.

#This heatmap reveals that the most significant relationship exists between GPA and GMAT scores, 
#suggesting that these two factors might be important predictors for some outcome (e.g., admission to a program). 
#The other variables seem to have little to no impact on each other.
```

```python
# 8. Boxplot to compare values within groups
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
sns.boxplot(data=data, x='gender', y='gpa', palette='pastel')
plt.title('GPA Distribution by Gender')
plt.show()

# Interpretation: Boxplot illustrates GPA variations across gender, showing central tendencies and outliers.
'''The box plot illustrates the distribution of GPAs by gender. Both genders have a similar range of GPAs, 
with the majority of GPAs falling between 3.2 and 3.6 for both females and males. 
However, there are a few outliers for females with GPAs slightly above 3.6 and one outlier for males with a GPA close to 3.8.
Overall, the plot suggests that there isn't a substantial difference in GPA distribution between males and females in this dataset. 
 '''
```

```python
# 9. Violin plot to compare distributions
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("MBA.csv")
sns.violinplot(data=data, x='work_industry', y='work_exp', palette='muted')
plt.title('Work Experience Distribution by Industry')
plt.xticks(rotation=45)
plt.show()

# Interpretation: Violin plot combines boxplot and KDE, revealing the density and spread of work experience across industries.
'''This violin plot illustrates the distribution of work experience (work_exp) across various industries. We can see that the 
median work experience varies significantly between industries. For example, the median work experience in Financial Services 
is around 6 years, while in Consulting, it's closer to 8 years. The spread of work experience also differs across industries. 
In some industries like Investment Banking, there's a wider range of work experience, with some individuals having less than 4 years 
and others having over 10. In contrast, industries like Consulting and Nonprofit/Gov have a more concentrated distribution, with most 
individuals having between 6 and 8 years of experience. Overall, this plot highlights the diverse work experience profiles across different industries.
 '''
```

```python
import pandas as pd

# Example dataset
data = pd.DataFrame({
    'gender': ['Female', 'Male', 'Female', 'Male', 'Male'],
    'major': ['Business', 'Humanities', 'Business', 'STEM', 'STEM']
})

# One-Hot Encoding
one_hot_encoded = pd.get_dummies(data, columns=['gender', 'major'], drop_first=False)
print(one_hot_encoded)
```

```python
# Dummy Encoding
dummy_encoded = pd.get_dummies(data, columns=['gender', 'major'], drop_first=True)
print(dummy_encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Ordinal Encoding
ordinal_encoder = OrdinalEncoder(categories=[['Female', 'Male']])
data['gender_encoded'] = ordinal_encoder.fit_transform(data[['gender']])
print(data)
```

```python
import pandas as pd
from category_encoders import BinaryEncoder

# Example dataset
data = pd.DataFrame({
    'major': ['Business', 'Humanities', 'Business', 'STEM', 'STEM']
})

# Binary Encoding
binary_encoder = BinaryEncoder(cols=['major'])
binary_encoded = binary_encoder.fit_transform(data)
print(binary_encoded)
```

```python
# Count Encoding
data['major_count_encoded'] = data['major'].map(data['major'].value_counts())
print(data)
```

```python
# Example dataset with target variable for Target Encoding
data = pd.DataFrame({
    'major': ['Business', 'Humanities', 'Business', 'STEM', 'STEM'],
    'admission': [1, 0, 1, 0, 0]
})

# Target Encoding
target_mean = data.groupby('major')['admission'].mean()
data['major_target_encoded'] = data['major'].map(target_mean)
print(data)
```
